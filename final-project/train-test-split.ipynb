# Nasdaq

Use to split the training and testing NASDAQ data for training model purpose. It will save 2 file: 1 contains data before the filter date and 1 contains data after the filter date

```python
#change the working path to nasdaq folder
path = os.chdir('/content/drive/MyDrive/Spring 2023/Deep Learning/Final Project/nasdaq')
```

```python
filter_date = datetime(2017, 3, 13) #choose the desire date to split
```

```python
#split the data file into 2 subfiles
for filename in os.listdir(path): #lopp through all the files inside folder nasdaq
  if filename.endswith(".csv"): #only work with csv file
    #create filename to save the file
    train_filename = filename[:-4] + '_train' 
    test_filename = filename[:-4] + '_test'

    #prepare the dataframe to split
    df = pd.read_csv(filename)
    df['Date']= pd.to_datetime(df['Date'], format= '%d-%m-%Y') #change the date column to datetime format

    #split the dataframe into 2 dataframes
    train = df[df['Date'] < filter_date]
    test = df[df['Date'] > filter_date]
    train = train.reset_index(drop=True)
    test = test.reset_index(drop=True)

    #save the 2 new dataframes to the desire folders
    train.to_csv("/content/drive/MyDrive/Spring 2023/Deep Learning/Final Project/nasdaq/train data/" + train_filename)
    test.to_csv("/content/drive/MyDrive/Spring 2023/Deep Learning/Final Project/nasdaq/test data/" + test_filename)
```

# VN

```python
#change the working path to vn folder
path = os.chdir('/content/drive/MyDrive/Spring 2023/Deep Learning/Final Project/vn')
```

```python
filter_date = datetime(2020, 3, 13) #choose the desire date to split
```

```python
#split the data file into 2 subfiles
for filename in os.listdir(path): #lopp through all the files inside folder nasdaq
  if filename.endswith(".csv"): #only work with csv file
    #create filename to save the file
    train_filename = filename[:3] + '_train' 
    test_filename = filename[:3] + '_test'

    #prepare the dataframe to split
    df = pd.read_csv(filename)
    df.rename(columns = {'TradingDate':'Date'}, inplace=True)
    df['Date']= pd.to_datetime(df['Date'], format= '%Y-%m-%d') #change the date column to datetime format

    #split the dataframe into 2 dataframes
    train = df[df['Date'] < filter_date]
    test = df[df['Date'] > filter_date]
    train = train.reset_index(drop=True)
    test = test.reset_index(drop=True)

    #save the 2 new dataframes to the desire folders
    train.to_csv("/content/drive/MyDrive/Spring 2023/Deep Learning/Final Project/vn/train data/" + train_filename)
    test.to_csv("/content/drive/MyDrive/Spring 2023/Deep Learning/Final Project/vn/test data/" + test_filename)
```

# Testing file nasdaq

Use to split the testing NASDAQ data for testing model purpose. It will save 1 file that contains data after the filter date.

```python
#change the working path to nasdaq folder
path = '/content/drive/MyDrive/Spring 2023/Deep Learning/Final Project/nasdaq/different industry'
```

```python
filter_date = datetime(2017, 3, 13) #choose the desire date to split
```

```python
#split the data file into 2 subfiles
for filename in os.listdir(path): #lopp through all the files inside folder nasdaq
  if filename.endswith(".csv"): #only work with csv file
    print(filename)
    #prepare the dataframe to split
    df = pd.read_csv(filename)
    df['Date']= pd.to_datetime(df['Date'], format= '%d-%m-%Y') #change the date column to datetime format
    # Filter the dataframe based on the filter_date
    df = df[df['Date'] > filter_date]
    df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')  # Change the date format to Y-M-D
    # Replace the file with the new file in the same folder
    new_filename = os.path.join(path, 'new'+filename)  # Create the new filename with the same path
    df.to_csv(new_filename, index=False)  # Save the dataframe as a new CSV file
```
